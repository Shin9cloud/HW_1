```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://20191118.tistory.com/3?category=761882')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select_one('span > b')
print(items)
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['번호','용어','내용'])
excel_sheet.column_dimensions['A'].width = 20
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 80
excel_sheet.column_dimensions['D'].width = 80

res = requests.get('https://20191118.tistory.com/3?category=761882')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('div > p')

# title = items[0].select_one('b')
# print(title.get_text())
for index, i in enumerate(range(10)):
    if items[i].get_text() == '':
        pass
    else:
        content = items[i]
        print(index, content.get_text())
        excel_sheet.append([index+1, content.get_text()])
        excel_sheet.cell(row = index+2, column=4)
excel_file.save('금융용어.xlsx')
excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G06')
soup = BeautifulSoup(res.content, 'html.parser')
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
items = bestitems.select('ul > li')

for item in items:
    title = item.select_one('a.itemname')
    price = item.select_one('div.s-price > strong')
    print(title) ; print(price)
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G06')
soup = BeautifulSoup(res.content, 'html.parser')
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
items = bestitems.select('ul > li')

for item in items:
    title = item.select_one('a.itemname')
    price = item.select_one('div.s-price > strong')
    print(title.get_text(), price.get_text(), title['href'])
```
